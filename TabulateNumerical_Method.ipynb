```python
import sympy
import numpy as np
from numpy import sin, cos, log, exp, sqrt, pi
from tabulate import tabulate


def cari_aproksimasi_akar():                                                    #pendefinisian program
    print("Akan dicari aproksimasi akar dengan metode Bisection, Fixed Point, Newton, Newton dengan beda hingga, Secant, Regula Falsi, dan Steffensen")


    ulang_program = 'ya'                                                        #untuk mengulang program
    while ulang_program =="ya":                                                 #ketika "ya" maka program akan diulang
        x = sympy.symbols("x")                                                  #pendefinisian simbol x sebagai variabel
        t = sympy.symbols("t")                                                  #pendefinisian simbol t sebagai variabel
        fungsi = input("Masukkan fungsi f(x) dengan variabel t: ")              #fungsi yang ingin dihitung
        NPM = eval(input("Masukkan NPM kamu: "))                                #input NPM untuk mencari nilai t
        def f(x,t):                                                             #pendefinisian fungsi terhadap variabel x dan t
            return eval(fungsi)

        t = NPM%2                                                               #Mendefinisikan t=0 genap atau t=1 ganjil dengan sisa modulus 2
        print("Maka nilai t kamu adalah = ",t)                                  #Akan diberitahukan nilai t


        print("Isi interval [a,b] ya! (boleh bergantung t)")
        lower = eval(input("Masukkan batas bawah interval (a): "))              #Input batas bawah
        upper = eval(input("Masukkan batas atas interval (b): "))               #Input batas atas
        print("Interval yang akan kamu gunakan adalah [",lower,",",upper,"]")


        starting = eval(input("Tebakan awal kamu adalah "))
        tolerance = eval(input("Kamu menggunakan toleransi aproksimasinya:"))
        iterationmax = eval(input("Masukkan batas maksimal iterasi: "))

        bisection = []                                                          #Pencarian aproksimasi akar metode bisection
        bisection_low=lower
        bisection_up=upper
        if f(bisection_low,t)*f(bisection_up,t)<0:                              #Syarat Utama pembaruan untuk variabel batasan bawah
          p0=bisection_low                                                      #Nilai a(bisection_low) tidak berubah, b(bisection_up) berubah menjadi p
          p=(bisection_low+bisection_up)/2                                      #Didefinisikan dari rumus nilai tengah interval p = (a+b)/2
          bisection.append(p)

          if f(p,t)==0:
            bisection.append("Selesai")                                         #Saat telah mencapai batas toleransi

          else:
              if f(p,t)*f(bisection_low,t)>0:
                bisectionlow=p
              elif f(p,t)*f(bisection_up,t)<0:
                bisection_up=p

              abs_error=abs(p0-p)
              p0=p

              while abs_error > tolerance:
                    p=(bisection_low+bisection_up)/2
                    bisection.append(p)

                    if f(p,t)==0:
                      break                                                     #Pemutusan loop
                    elif f(p,t)*f(bisection_low,t)>0:
                      bisection_low=p                                           #Pembaruan variabel batas bawah
                    elif f(p,t)*f(bisection_low,t)<0:
                      bisection_up=p                                            #Pembaruan variabel batas atas

                    abs_error=abs(p0-p)                                         #Pembaruan value variabel loop
                    p0=p

        else:
          if f(bisection_low,t)*f(bisection_up,t)>0:                            #Syarat ke-2 pembaruan variabel untuk batas atas
              for i in range(iterationmax):
                bisection.append(np.nan)                                        #Metode gagal terhadap batas atas dan bawah yang diinput
          else:
              if f(bisection_low,t)==0:
                  bisection.append(bisection_low)
              else:
                  bisection.append(bisection_up)

        fixed_point=[]                                                          #Metode Fixed Point
        p0=starting                                                             #titik awal iterasi
        p=p0-f(p0,t)                                                            #Didapatkan dari g(x)=x-f(x)
        fixed_point.append(p)
        for i in range(iterationmax):                                           #Akan terus looping hingga batas iterasi
            if abs(p-p0)>=tolerance:                                            #Pembaruan value variabel loop
                p0=p
                p=p0-f(p0,t)
                fixed_point.append(p)

        newton_biasa=[]                                                         #Metode Newton Dengan Turunan Analitik
        df_string = str(sympy.diff(fungsi, x))
        def df(x,t): # turunan f
            return eval(df_string)

        p0=starting
        p = p0 - f(p0,t)/df(p0,t)                                               #Didapatkan dari G(x)=x-f(x)/f'(x)
        abs_error = abs(p-p0)
        p0 = p
        newton_biasa.append(p)

        while abs_error > tolerance:
            try:
                p = p0 - f(p0,t)/df(p0,t)
            except ZeroDivisionError:
                for i in range(iterationmax):
                    newton_biasa.append(np.nan)                                 #Metode akan gagal (nan) ketika turunanya = 0
                break                                                           #Stop Loop
            abs_error = abs(p-p0)                                               #Pembaruan value variabel loop
            p0 = p
            newton_biasa.append(p)

        newton_beda_hingga=[]                                                   #Metode Newton Dengan Beda Hingga
        def df(x,t, h=10**(-12)):                                               #Diketahui dalam tugas h=10**(-12)
            return (f(x+h,t)-f(x,t))/h                                          #Mencari turunan

        p0=starting
        p=p0-(f(p0,t)/df(p0,t))
        newton_beda_hingga.append(p)
        for i in range(iterationmax):
            if abs(p-p0)>=tolerance:
                p0=p
                p=p0-(f(p0,t)/df(p0,t))
                newton_beda_hingga.append(p)

        regula_falsi = []                                                       #Metode Regula Falsi
        regulalb=lower                                                          #Mendefinisikan batas bawah
        regulaub=upper                                                          #Mendefinisikan batas atas
        if f(regulalb,t)*f(regulaub,t)<0:
            p0=regulalb
            p1=regulaub
            p=p1-(f(p1,t)*(p1-p0))/(f(p1,t)-f(p0,t))                            #dari perubahan rumus metode bisection
            regula_falsi.append(p)

            if f(p,t)==0:
                regula_falsi.append("Selesai")                                  #telah mencapai batas toleransi
            else:
                if f(p,t)*f(regulalb,t)>0:
                    regulalb=p
                elif f(p,t)*f(regulalb,t)<0:
                    regulaub=p

                abs_error=abs(p0-p)
                p0=p

                while abs_error > tolerance:
                    p=p1-(f(p1,t)*(p1-p0))/(f(p1,t)-f(p0,t))
                    regula_falsi.append(p)
                    if f(p,t)==0:
                        break
                    elif f(p,t)*f(regulalb,t)>0:
                        regulalb=p
                    elif f(p,t)*f(regulalb,t)<0:
                        regulaub=p

                    abs_error=abs(p0-p)
                    p0=p

        else:
            if f(regulalb,t)*f(regulaub,t)>0:
                for i in range(iterationmax):
                    regula_falsi.append(np.nan)
            else:
                if f(regulalb,t)==0:
                    regula_falsi.append(regulalb)
                else:
                    regula_falsi.append(regulaub)

        secant=[]                                                               #Metode Secant
        p0=1-0.5*t                                                              #Titik awal atau batas bawah
        p1=4.5                                                                  #Titik kedua atau batas atas
        p=p1-(f(p1,t)*(p1-p0))/(f(p1,t)-f(p0,t))                                #variabel atau rumus untuk mencari abs_errornya
        secant.append(p)
        for i in range(iterationmax):
            if abs(p-p1)>=tolerance:                                            #variabel validasi akan loop selama nilainya >= dari toleransi
                p0=p1                                                           #Pembaruan value untuk validasi loop
                p1=p                                                            #Pembaruan value untuk validasi loop
                p=p1-(f(p1,t)*(p1-p0))/(f(p1,t)-f(p0,t))
                secant.append(p)

        steffensen = [starting]                                                 #Metode Steffensen
        abs_error = tolerance + 1                                               #Definisi pembaruan value variabel loop
        iterasi = 1                                                             #Iterasi dimulai dari 1

        while abs_error >= tolerance and iterasi <=51:                          #Akan terus looping ketika value variabel >= batas loreansi dan iterasinya <=51
            if iterasi % 3 == 0:                                                #Menggunakan Aitken untuk kelipatan tiga
                pn_3 = steffensen[iterasi - 3]
                pn_2 = steffensen[iterasi - 2]
                pn_1 = steffensen[iterasi - 1]
                Delta1 = pn_2 - pn_3
                Delta2 = pn_1 - 2 * pn_2 + pn_3
                pn = pn_3 - (Delta1)**2 / Delta2

            else:                                                               #Selain kelipatan tiga, akan digunakan fixed point
                pn_1 = steffensen[iterasi - 1]
                pn = pn_1-f(pn_1,t)

            steffensen.append(pn)                                               #return pn
            abs_error = abs( pn - pn_1 )
            iterasi+=1

        del steffensen[0]

        method = [bisection, fixed_point, newton_biasa, newton_beda_hingga, secant, regula_falsi, steffensen] #Mendefinisikan metode
        max_iterationmax=0
        for i in range(len(method)):
            if np.isnan(sum(method[i]))== False:                                #Jika data bukan berupa nan
                for j in range(len(method[i])):
                    method[i][j]="{:.9f}".format(method[i][j])

                if len(method[i])>max_iterationmax:                             #looping hingga batas iterasinya
                    max_iterationmax=len(method[i])

        for i in range(len(method)):
            if len(method[i])>max_iterationmax:
                method[i]=method[i][:max_iterationmax]
            else:
                while len(method[i])<max_iterationmax:                          #Looping akan mengeluarkan output "Selesai" jika sudah melewati batas toleransi
                    method[i].append("Selesai")

        iterasi=[]
        for i in range(max_iterationmax):
            iterasi.append(str(i+1))                                            #looping iterasi hingga iterasi maksimal

        headers=['Number',"Bisection","Fixed Point","Newton","Newton Beda Hingga","Secant","Regula Falsi","Steffensen"] #Pemberian headers
        table=tuple(zip(iterasi, bisection, fixed_point, newton_biasa, newton_beda_hingga, secant, regula_falsi, steffensen))
        print(tabulate(table,headers,tablefmt='grid'))                          #Data dimasukkan ke dalam rumus tabel

        ulang_program=input("Apakah kamu mau ulangi program ini? ketik 'ya' untuk mengulanginya! ") #Pertanyaan ulang program

        if ulang_program=="ya":                                                 #Formula akan mengulang program jika diberikan input "ya"
            print('Baik, program akan diulang!')

    print("Terima kasih telah menggunakan program ini!")                        #Jika diberikan input selain "ya" maka program akan berhenti dan mem-print kata kata di samping.
```
