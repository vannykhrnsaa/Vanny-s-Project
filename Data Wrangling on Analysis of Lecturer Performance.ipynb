# **Task 01:**

Apakah ada dosen yang cenderung memberikan nilai cenderung lebih tinggi (atau lebih rendah) dibandingkan dosen lainnya?

```python
# Code Task 01

# membuat table nilai rata-rata masing-masing kolom table nilai
nilai_dosen = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1)

# merge left table dosen dan nilai_dosen dengan key: NIDN
nilai_dosen = pd.merge(dosen, nilai_dosen, left_on='NIDN', right_index=True)
nilai_dosen = nilai_dosen[['NIDN', 'nama', 'mean_nilai']]

# membuat table nilai standar deviasi masing-masing kolom table nilai
sb = nilai.groupby(nilai['NIDN']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1)

# concat nilai_dosen dan sb
nilai_dosen = nilai_dosen.set_index(['NIDN'])
nilai_dosen = pd.concat([nilai_dosen, sb], axis=1)
nilai_dosen
```

```python
# Code Task 02

# merge table dosen dan nilai dengan key: NIDN
merged_data = pd.merge(dosen, nilai, on='NIDN')
result_data = merged_data[['Pendidikan_Terakhir', 'nilai']]

# membuat table nilai rata-rata masing-masing kolom table result_data
mean_data = result_data.groupby(merged_data['Pendidikan_Terakhir']).mean()[['nilai']]
mean_data = mean_data.rename({'nilai': 'mean_nilai'}, axis=1)

# membuat table nilai standar deviasi masing-masing kolom table merge_data
std_data = merged_data.groupby(merged_data['Pendidikan_Terakhir']).std()[['nilai']]
std_data = std_data.rename({'nilai': 'std_nilai'}, axis=1)

# concat table mean_data dan std_data
final_data = pd.concat([mean_data, std_data], axis=1)
final_data
```

```python
# Code Task 03

# buat table baru yang komponennya nilai kolom nim dan nilai dari table nilai
nilai_extracted = nilai[['nim','nilai']]

# buat table baru yang komponennya nilai kolom nim dan asal sekolah dari table mahasiswa
mahasiswa_extracted = mahasiswa[['nim', 'Asal_Sekolah']]

# merge left table nilai_extracted dan mahasiswa_extracted dengan key: nim
merged_data = pd.merge(mahasiswa_extracted, nilai_extracted, on='nim', how='left')

# membuat table nilai rata-rata masing-masing kolom table merged_data
mean_data = merged_data.groupby('Asal_Sekolah')[['nilai']].agg('mean')
mean_data = mean_data.rename({'nilai': 'mean_nilai'}, axis=1)

# membuat table nilai standar deviasi masing-masing kolom table merged_data
std_data = merged_data.groupby(merged_data['Asal_Sekolah']).std()[['nilai']]
std_data = std_data.rename({'nilai': 'std_nilai'}, axis=1)

# concat table mean_data dan std_data
final_data = pd.concat([mean_data, std_data], axis=1)
final_data
```

## **Code 06 #2**

```python
# Code Soal Untuk No 06 #2
nilai_extracted = nilai[['id_mata_kuliah','nilai']]

# membuat table nilai rata-rata masing-masing kolom table nilai_extracted
mean_data = nilai_extracted.groupby('id_mata_kuliah')[['nilai']].agg('mean')
mean_data = mean_data.rename({'nilai': 'mean_nilai'}, axis=1)

# membuat table nilai standar deviasi masing-masing kolom table nilai_extracted
std_data = nilai_extracted.groupby(nilai_extracted['id_mata_kuliah']).std()[['nilai']]
std_data = std_data.rename({'nilai': 'std_nilai'}, axis=1)

# concat
combined_data = pd.concat([mean_data,std_data], axis=1)

matkul_df = pd.DataFrame(nilai['id_mata_kuliah'].value_counts())
matkul_df = matkul_df.rename({'id_mata_kuliah': 'frekuensi_mahasiswa_per_kelas'}, axis=1)
matkul_df.reset_index(inplace=True)
matkul_df = matkul_df.rename(columns = {'index':'matkul'})

matakuliah_df = pd.merge(combined_data, matkul_df, left_on='id_mata_kuliah', right_on='matkul')
matakuliah_df = matakuliah_df.set_index('matkul')
matakuliah_df
```
