```python
import random
# Inisialisasi list untuk menyimpan semua iterasi
iterasi_data = []

# Jumlah iterasi yang diinginkan
jumlah_iterasi = 7

for iterasi in range(jumlah_iterasi):
    # Membuat DataFrame iterasi untuk kolom 'x' dan 'y'
    iterasi_df = pd.DataFrame({
        'x': x,
        'y': (iterasi + 2) # Nilai 'y' sesuai dengan iterasi saat ini
    })

    # Mengisi kolom lainnya dengan nilai random sesuai dengan interval yang diberikan
    iterasi_df['Corn97BU'] = [round(random.uniform(150.00, 199.00), 2) for _ in range(len(x))]
    iterasi_df['B'] = [round(random.uniform(1.00, 5.99), 2) for _ in range(len(x))]
    iterasi_df['Ca'] = [random.randint(1500, 3999) for _ in range(len(x))]
    iterasi_df['Cu'] = [round(random.uniform(0.50, 2.60), 2) for _ in range(len(x))]
    iterasi_df['Fe'] = [random.randint(100, 299) for _ in range(len(x))]
    iterasi_df['K'] = [random.randint(60, 199) for _ in range(len(x))]
    iterasi_df['Mg'] = [random.randint(250, 650) for _ in range(len(x))]
    iterasi_df['Mn'] = [random.randint(40, 120) for _ in range(len(x))]
    iterasi_df['Na'] = [random.randint(11, 19) for _ in range(len(x))]
    iterasi_df['P'] = [random.randint(8, 49) for _ in range(len(x))]
    iterasi_df['Zn'] = [round(random.uniform(0.30, 3.20), 2) for _ in range(len(x))]

    # Menambahkan DataFrame iterasi ke list iterasi_data
    iterasi_data.append(iterasi_df)

data_tambahan = pd.concat(iterasi_data, ignore_index=True)
pd.set_option('display.max_rows', None)
# Menampilkan hasil
print(data_tambahan)
```

```python
data_lengkap = pd.concat([data, data_tambahan], ignore_index=True)
pd.set_option('display.max_rows', None)
# Menampilkan hasil penggabungan
print(data_lengkap)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Menghitung logaritma dari kolom 'Ca'
data_lengkap['log10(Ca)'] = np.log10(data_lengkap['Ca'])

# Membuat dotplot dengan kondisi untuk marker style
plt.figure(figsize=(8, 4))
for i, log_ca in enumerate(data_lengkap['log10(Ca)']):
    if log_ca <= 3.55:
        marker_style = 'bo'
    else:
        marker_style = 'rx'
    plt.plot(log_ca, i, marker_style, markersize=8, alpha=0.5)

plt.xlabel('log10(Ca)')
plt.ylabel('Index')
plt.title('Dotplot Distribusi log10(Ca)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Menghitung logaritma dari kolom 'Ca'
data_lengkap['log10(Fe)'] = np.log10(data_lengkap['Fe'])
data_lengkap['log10(Corn97BU)'] = np.log10(data_lengkap['Corn97BU'])
data_lengkap['log10(Ca)'] = np.log10(data_lengkap['Ca'])  # Menambahkan ini untuk log10(Ca)

# Membuat data contoh
x = data_lengkap['log10(Fe)']
y = data_lengkap['log10(Corn97BU)']

# Inisialisasi warna dan style marker
colors = np.where(data_lengkap['log10(Ca)'] <= 3.55, 'blue', 'red')  # Menggunakan 'red' untuk nilai log10(Ca) > 3.55

# Membuat scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=colors, marker='o', label='Data Points', alpha=0.5)  # Menggunakan 'colors' untuk warna

# Mengidentifikasi indeks yang akan menggunakan marker 'x' berdasarkan nilai log10(Ca) > 3.55
indices_to_mark_as_x = np.where(data_lengkap['log10(Ca)'] > 3.55)[0]
for index in indices_to_mark_as_x:
    plt.scatter(x.iloc[index], y.iloc[index], marker='x', c='red', s=100)  # Marker 'x' berwarna merah

plt.title('Scatterplot: Hubungan antara log10(Fe) dan log10(Corn97BU)')
plt.xlabel('log10(Fe)')
plt.ylabel('log10(Corn97BU)')

# Menambahkan legenda
plt.legend()

plt.grid(True)  # Menambahkan grid (opsional)

plt.show()
```

##Stripplot masing-masing elemen

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat DataFrame Anda (misalnya, data_lengkap)
# Pastikan data_lengkap berisi kolom-kolom yang sesuai (B, Ca, Cu, Fe, K, Mg, Mn, Na, P, Zn)

# Menghitung logaritma basis 10 dari kolom 'Ca'
data_lengkap['log10(Ca)'] = np.log10(data_lengkap['Ca'])

# List kolom yang ingin Anda buat stripplot-nya
kolom_stripplot = ['B', 'Ca', 'Cu', 'Fe', 'K', 'Mg', 'Mn', 'Na', 'P', 'Zn']

# Membuat stripplot untuk setiap kolom
plt.figure(figsize=(15, 10))  # Menentukan ukuran figur

for i, kolom in enumerate(kolom_stripplot, 1):
    plt.subplot(2, 5, i)
    sns.stripplot(x=kolom, y=np.random.rand(len(data_lengkap)), data=data_lengkap, jitter=True, alpha=0.7)
    plt.xlabel(kolom)
    plt.ylabel('')
    plt.title(f'Stripplot {kolom} Jitter Horizontal')

plt.tight_layout()
plt.show()
```
